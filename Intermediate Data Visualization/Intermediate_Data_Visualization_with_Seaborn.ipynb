## Data Visualization

### Histogram plotting function of pandas library

```
df['Tuition'].plot.hist()
```

### Background Theme Examples

```
for style in ['white','whitegrid','dark','darkgrid','ticks']:
  sns.set_style(style)
  sns.distplot(df['Tuition'])
  plt.show()
```

### Removing spines from the plot

```
sns.distplot(df['Tuition'])
sns.despine(left=True)
plt.show()
```

### Changing the color of the plot using Matplotlib colors

```
sns.distplot(df['HIGHDEG'], color='r')
plt.show()
```

### Seaborn Palletes

```
for palette in sns.palettes.SEABORN_PALETTES:
  sns.set_palette(palette)
  sns.distplot(df['CONTROL'])
  plt.show()
```

#### Various Seaborn palettes

```
for p in sns.palettes.SEABORN_PALETTES:
  sns.set_palette(p)
  sns.palplot(sns.color_palette())
  plt.show()
```

```
sns.palplot(sns.color_palette("BrBG", 12))
```

```
sns.palplot(sns.color_palette("Reds", 12))
```

### Bar plot using Catplot() function of Seaborn library

```
sns.catplot(data=df, x="Degree_Type" ,y="ICLEVEL", kind='bar')
plt.show()
```

### Bar plot with 3rd parameter

```
sns.catplot(data=df, x="Degree_Type" ,y="ICLEVEL", kind='bar',hue="CONTROL")
plt.show()
```

```
sns.catplot(data=df, x="Degree_Type" ,y="HIGHDEG", kind='bar')
plt.show()
```

### Box plot using catplot() function of Seaborn library

```
sns.catplot(data=df, x="Degree_Type" ,y="Tuition", kind='box')
plt.show()
```

### KDE- Kernel Density Plot

```
sns.distplot(df['ICLEVEL'], color='g',kde=True)
plt.show()
```

```
sns.distplot(df['ICLEVEL'],color='g',kde=False)
plt.show()
```
