## Importing the required libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset

## Exploratory Data Analysis

```
df = pd.read_csv('college_data.csv')
```

```
df.head()
```

```
df.isna().sum()
```

## Data Visualization

### Plotting with regplot() and evaluating regression with residplot() function

```
fig, [ax0, ax1] = plt.subplots(1, 2)
fig.set_size_inches([12,6])
sns.regplot(data=df, x="Tuition", y="SAT_AVG_ALL", ax=ax0)
sns.residplot(data=df, x="Tuition", y="SAT_AVG_ALL", ax=ax1)
plt.show()
```

### Polynomial Regression with regplot() and residplot()

```
fig, [ax0, ax1] = plt.subplots(1, 2)
fig.set_size_inches([12,6])
sns.regplot(data=df, x="Tuition", y="SAT_AVG_ALL", order=2, ax=ax0)
sns.residplot(data=df, x="Tuition", y="SAT_AVG_ALL", order=2, ax=ax1)
plt.show()
```

### Binning the data

```
fig, [ax0, ax1] = plt.subplots(1,2)
fig.set_size_inches([14,6])
sns.regplot(data=df, x="Tuition", y="PCTPELL",  ax=ax0)
sns.regplot(data=df, x="Tuition", y="PCTPELL", x_bins=10, ax=ax1)
plt.show()
```
