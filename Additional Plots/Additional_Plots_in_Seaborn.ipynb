## Data Visualization : Categorical Plot Types

### Plot types- show each observations :

#### Strip-Plot

```
sns.stripplot(data=df, x="Region", y="Award_Amount")
plt.show()
```

```
sns.stripplot(data=df,
         y='Award_Amount',
        x='Region', jitter=False)

plt.show()
```

#### Swarm-Plot

```
fig, [ax0, ax1] = plt.subplots(1,2)
fig.set_size_inches([18,6])
sns.swarmplot(data=df, x="Award_Amount", y="Model Selected", hue="Region", dodge=False, ax=ax0)
sns.swarmplot(data=df, x="Award_Amount", y="Model Selected", hue="Region", dodge=True, ax=ax1)
sns.set_style('dark')
plt.show()
```

### Plot types- show abstract representations

#### Boxplot, Violinplot and Lvplot/Boxenplot

```
fig,[ax0, ax1, ax2] = plt.subplots(1,3, sharey=True)
fig.set_size_inches([18,6])
sns.boxplot(data=df, y="Award_Amount", x="Model Selected", ax=ax0)
sns.violinplot(data=df, y="Award_Amount", x="Model Selected", ax=ax1)
sns.boxenplot(data=df, y="Award_Amount", x="Model Selected", ax=ax2)
plt.show()
```

#### Boxplot, Violinplot and Lvplot/Boxenplot with 3rd parameter and different palettes

```
fig,[ax0, ax1, ax2] = plt.subplots(3,1, sharex=True)
fig.set_size_inches([15,15])
sns.boxplot(data=df, y="Award_Amount", x="Model Selected",hue="Region", ax=ax0, palette="pastel")
sns.violinplot(data=df, y="Award_Amount", x="Model Selected",hue="Region", ax=ax1, palette="Purples")
sns.boxenplot(data=df, y="Award_Amount", x="Model Selected",hue="Region", ax=ax2, palette="Paired")
plt.show()
```

### Plot types- Statistical estimates

#### Bar-plot

```
sns.catplot(data=df, y="Model Selected", x="Award_Amount", hue="Region" ,kind='bar')
plt.show()
```

#### Point-plot

```
sns.catplot(data=df, kind='point', y="Award_Amount", x="Model Selected", col="Region")
plt.show()
```

```
sns.catplot(data=df, kind='point', y="Award_Amount", x="Model Selected", row="Region")
plt.show()
```

#### Count-plot

```
sns.catplot(data=df, kind='count', x="Region", hue="Model Selected")
plt.show()
```
