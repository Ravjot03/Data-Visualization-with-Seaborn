## Importing the libraries

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Reading the dataset

```
df = pd.read_csv('student-alcohol-consumption.csv')
df.head()
```

# Data Visualization Part

### Using **sns.relplot** -- Relational Plot

```
sns.relplot(data=df, x='absences', y='G3',kind='scatter')
plt.show()
```

## Subplots in Columns

```
sns.relplot(x="absences", y="G3", data=df,kind="scatter",col="study_time")
plt.show()
```

## Subplots in Rows

```
sns.relplot(x="absences", y="G3", data=df,kind="scatter",row="study_time")
plt.show()
```

```
sns.relplot(x="G1", y="G3", data=df, kind='scatter')
plt.show()
```

## Ordering Columns

```
sns.relplot(x="G1", y="G3", data=df, kind="scatter",col="schoolsup", col_order=["yes","no"])
plt.show()
```

## Subplots in Rows and Columns and Ordering them

```
sns.relplot(x="G1", y="G3", 
            data=df,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],row="famsup",
            row_order=["yes", "no"])
plt.show()
```

## Reading the 2nd dataset

```
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

## Changing the size of scatter plot points

```
sns.relplot(x="horsepower",y="mpg",size="cylinders",data=mpg, kind='scatter')
plt.show()
```

## Setting the Style and Color of the Scatter plot

```
sns.relplot(x='acceleration',y='mpg', data=mpg, kind='scatter',style="origin",hue="origin")
plt.show()
```

## Line Plot

```
sns.relplot(data=mpg,x="model_year",y="mpg",kind='line')
plt.show()
```

## Visualizing Standard Deviation with Line Plot

```
sns.relplot(data=mpg,x="model_year",y="mpg",kind='line', ci='sd')
plt.show()
```

## Plotting subgroups in Line Plots

```
sns.relplot(data=mpg, x="model_year",y="horsepower", kind='line',ci=None)
plt.show()
```

## Setting the style and color of the Line plot

```
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None,style="origin",hue="origin")
plt.show()
```

## Setting the Markers in the Line plot and making each line with the same style

```
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",markers=True,dashes=False)
plt.show()
```
