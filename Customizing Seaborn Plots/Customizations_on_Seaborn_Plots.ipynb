## Data Visualization

### Regression Plot using regplot() function of Seaborn library

```
sns.regplot(data=df, x="acs_2017_2", y="fmr_0")
plt.show()
```

```
sns.regplot(data=df, x="pop2010", y="fmr_1")
plt.show()
```

### Regression plot with Scatter parameter as **False**

```
sns.regplot(data=df, x="pop2010", y="fmr_1", scatter=False)
plt.show()
```

### Regression Plot with different **orders**

```
#fig.set_size_inches([10,10])
fig, [ax0, ax1] = plt.subplots(1, 2)
fig.set_size_inches([14,4])

sns.regplot(data=df, x="pop2010", y="fmr_1", order=0, ax=ax0)
sns.regplot(data=df, x="pop2010", y="fmr_1", order=2, ax=ax1)
sns.set_style('dark')
plt.show()
```

### lmplot - Seaborn : 
2D scatterplot with an optional overlaid regression line

```
sns.set_style('white')
# Create a regression plot
sns.lmplot(data=df,
           x='pop2010',
           y='fmr_2')
# Remove the spines
sns.despine(right=True, top=True)
# Show the plot and clear the figure
plt.show()
plt.clf()
```

### Different Palettes in plots

```
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    sns.distplot(df['fmr_3'])
    plt.show()
```

### Plot with title, xlabel functions

```
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of 1 bedroom rents
sns.distplot(df['fmr_1'], ax=ax)
# Modify the properties of the plot
ax.set(xlabel="1 Bedroom Fair Market Rent",
       xlim=(100,1500),
       title="US Rent")
# Display the plot
plt.show()
```

### Subplots for a better view of distribution

```
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=5, sharey=True)
fig.set_size_inches([15,4])
# Plot the distribution of 0 bedroom apartments on ax0
sns.distplot(df['fmr_0'], ax=ax0)
ax0.set(xlabel="0 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 1 bedroom apartments on ax1
sns.distplot(df['fmr_1'], ax=ax1)
ax1.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax2
sns.distplot(df['fmr_2'], ax=ax2)
ax2.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 3 bedroom apartments on ax3
sns.distplot(df['fmr_3'], ax=ax3)
ax3.set(xlabel="3 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 4 bedroom apartments on ax4
sns.distplot(df['fmr_4'], ax=ax4)
ax4.set(xlabel="4 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
```

### Customizations and combining plots

```
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=634.0, color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=706.3254351016984, color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
```

```
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=df['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)
ax.axvline(x=df['fmr_1'].std(), color='g', label='Standard Deviation', linestyle=':', linewidth=2)
# Show the legend and plot the data
ax.legend()
plt.show()
```
